```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def logistic_map(r, x):
    return r * x * (1 - x)

def logistic_map_iterated(r, x0, n):
    x = x0
    for i in range(n):
        x = logistic_map(r, x)
    return x

def logistic_map_list(r, x0, n):
    x = x0
    arr = []
    for i in range(n):
        arr.append(x)
        x = logistic_map(r, x)
    arr.append(x)
    return arr
```

```python
print(logistic_map_iterated(r=3.9, x0=0.7, n=10))
print(logistic_map_iterated(r=3.9, x0=0.701, n=10))
for i, x in enumerate(logistic_map_list(r=3.9, x0=0.7, n=10)):
    print(f"{i}: {x:.4f}")
```

```python
r = 3.9
plt.hist(logistic_map_list(r=3.9, x0=0.7, n=1000000), bins=1000, density=True)
plt.xlim(0, 1)
plt.yscale('log')
plt.show()
```

```python
hist_nested_list = []
for r in np.arange(2.0, 4.0, 0.005):
    h = np.histogram(logistic_map_list(r=r, x0=0.501, n=10000)[1000:], bins=500, range=(0.0, 1.0))[0]
    hist_nested_list.append(np.log(h + 0.1))
hist_nested_list = np.array(hist_nested_list).T
plt.figure(figsize=(12, 8))
plt.imshow(hist_nested_list, cmap='hot', interpolation='nearest', aspect='auto', origin='lower')
```
