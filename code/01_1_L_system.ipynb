# Example 1.1: L-System

```python
from typing import List

def update(letters: str) -> str:
    new_letters = ""
    for s in letters:
        if s == "R":
            new_letters += "RL"
        elif s == "L":
            new_letters += "R"
    return new_letters

def update_iterated(starting_string: str, n: int) -> str:
    output = starting_string
    for i in range(n):
        output = update(output)
    return output

def list_of_updates(starting_string: str, n: int) -> List[str]:
    output = []
    s = starting_string
    for i in range(n):
        output.append(s)
        s = update(s)
    output.append(s)
    return output

def get_RL_ratio(letters: str) -> float:
    num_L = 0.0
    num_R = 0.0
    for s in letters:
        if s == "R":
            num_R += 1
        elif s == "L":
            num_L += 1
    if num_L < 1:
        return 0
    return num_R/num_L
```

```python
progression = list_of_updates("R", 10)
for i, s in enumerate(progression):
    print(f"{i}: {s}")
```

```python
import matplotlib.pyplot as plt

ratios = [get_RL_ratio(s) for s in progression]
plt.plot(ratios, label="(# of R)/(# of L)")
plt.xlabel("Iterations")
plt.xlim(1, 10)
plt.ylim(1.0, 2.2)
plt.legend()
plt.show()
```
